```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.fftpack as sf
%matplotlib inline
size=256
scan_no=17
```

```python
obj = np.genfromtxt('example_data/scan%d_object_0.csv' %scan_no, delimiter=',', dtype=complex)
f, axarr=plt.subplots(1,2, figsize=(20,10))
img=axarr[0].imshow(np.abs(obj[50:-50,50:-50]))
f.colorbar(img,ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.angle(obj[50:-50,50:-50]))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```

```python
probes = np.genfromtxt('example_data/scan%d_probes_0.csv' %scan_no, delimiter=',', dtype=complex)
f,axarr=plt.subplots(1,2,figsize=(20,10))
img=axarr[0].imshow(np.abs(probes[:size].T))
plt.colorbar(img, ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.abs(probes[size:size*2].T))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```

```python
probes = np.genfromtxt('example_data/scan%d_probes_0.csv' %scan_no, delimiter=',', dtype=complex)
f,axarr=plt.subplots(1,2,figsize=(20,10))
img=axarr[0].imshow(np.angle(probes[:size].T))
plt.colorbar(img, ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.angle(probes[size:size*2].T))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```

```python
tmp1=sf.fftshift(sf.fftn(probes[:size]))
tmp2=sf.fftshift(sf.fftn(probes[size:size*2]))
f,axarr=plt.subplots(1,2,figsize=(20,10))
img=axarr[0].imshow(np.abs(tmp1.T[size/4:-size/4, size/4:-size/4]))
plt.colorbar(img, ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.abs(tmp2.T[size/4:-size/4, size/4:-size/4]))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```

```python
tmp1=sf.fftshift(sf.fftn(probes[:size]))
tmp2=sf.fftshift(sf.fftn(probes[size:size*2]))
f,axarr=plt.subplots(1,2,figsize=(20,10))
img=axarr[0].imshow(np.angle(tmp1.T[size/4:-size/4, size/4:-size/4]))
plt.colorbar(img, ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.angle(tmp2.T[size/4:-size/4, size/4:-size/4]))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```
