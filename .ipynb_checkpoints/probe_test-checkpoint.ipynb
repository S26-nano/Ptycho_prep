```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.fftpack as sf
import scipy.io as sio
%matplotlib inline
```

```python
data=sio.loadmat('example_data/probe_10.4kev_30_defocus.mat')
print (data.keys())
```

```python
probe=data['probe']
```

```python
f, axarr=plt.subplots(1,2, figsize=(20,10))
img=axarr[0].imshow(np.abs(probe))
f.colorbar(img,ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.angle(probe))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```

```python
f, axarr=plt.subplots(1,2, figsize=(20,10))
img=axarr[0].imshow(np.abs((sf.fftshift(sf.fftn(probe)))))
f.colorbar(img,ax=axarr[0], fraction=0.046, pad=0.04)
img=axarr[1].imshow(np.angle((sf.fftshift(sf.fftn(probe)))))
plt.colorbar(img, ax=axarr[1], fraction=0.046, pad=0.04)
```
